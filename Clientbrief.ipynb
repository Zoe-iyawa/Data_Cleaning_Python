```python
import pandas as pd.  #import panda library
```

```python
df = pd.read_excel('/Users/zoephilipiyawa/Downloads/Customer Call List.xlsx') 
# we want to read the file which is in excel format
df 
```

```python
df.drop_duplicates()  # remove duplicates
df = df.drop_duplicates()
```

```python
df.drop(columns = 'Not_Useful_Column')  # drop useless columns
df = df.drop(columns = 'Not_Useful_Column')
```

```python
df
```

```python
df['Paying Customer']= df['Paying Customer'].str.replace( 'No', 'N')   # standardise the columns
df['Paying Customer'] = df['Paying Customer'].str.replace( 'Yes', 'Y')
df['Paying Customer']
df
```

```python
df.isnull().sum()
```

```python
df['Do_Not_Contact']= df['Do_Not_Contact'].str.replace( 'No', 'N')  #standardize the columns 
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace( 'Yes', 'Y')
df['Do_Not_Contact']
df
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('/._') #strip the non alphabetical values from the last name column
```

```python
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','', regex = True) 
# remove the non-alphabetical symbols from Phone number column

```

```python
df
```

```python
df['Phone_Number']= df['Phone_Number'].str.replace('Na', ' ') # replace 'Na' with space from Phone number column
```

```python
df = df.fillna(' ') # fill the null values with blank spaces
```

```python
df[['Street_Address', 'State', 'Zip Code']]= df['Address'].str.split(',', expand = True) 
# split the address column and assign column names
```

```python
df
```

```python
for x in df.index:      # remove columns where clients do not want to be contacted. 
    if df.loc[x,'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace = True)
   
```

```python
for x in df.index:      # remove columns where clients have not put their phone number. 
    if df.loc[x,'Phone_Number'] == '':
        df.drop(x, inplace = True)
```

```python
df.dropna(subset = 'Phone_Number', inplace = 'True')
```

```python
df.drop(columns = 'Address', inplace = True) #drop the address column
```

```python
df
```

```python
df.reset_index(drop = True, inplace = True) # reset the column index 
```
